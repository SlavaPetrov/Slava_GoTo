```python
file_name = 'output.txt'

with open ("/Users/vvpetrov/Desktop/Python_Bio/small_ref.fa") as f:
    first = True
    refl = ''
    ref_name = ''
    for refline in f:
        if first :
            ref_name = refline.strip()
            first = False
            continue
        refl += line.strip()   
        
        
from pprint import pprint
def score(a, b):
    if a == '-' or b == '-':
        return -1
    
    if a == b:
        return 1
    
    if a != b:
        return -1

def align(s,t):
    a = []
    for i in range (0,len(s)+1):
        b = []
        for j in range (0,len(t)+1):
            b.append(0)
        a.append(b)

    for j in range (0, len(t)+1):
        a[0][j] = 0
    for i in range (0,len(s)+1):
        a[i][0] = 0
    
    for i in range (1,len(s)+1):
        for j in range (1,len(t)+1):
            x = a[i-1][j-1]+score(s[i-1],t[j-1])
            y = a[i-1][j]+score(s[i-1],'-')
            z = a[i][j-1]+score('-',t[j-1])
            a[i][j] = max(x,y,z,0)
    p = 0
    m = 0
    m1 = a[0][0]
    for i in range (1, len(s)+1):
        for j in range (1,len(t)+1):
            if a[i][j] >= m1:
                m1 = a[i][j]
                p = i
                m = j
    
    sa =''
    ta =''
     
    while a[p][m] != 0:
        if p > 0 and m > 0 and a[p][m] == a[p-1][m-1]+score(s[p-1],t[m-1]):
            sa = s[p-1] + sa
            ta = t[m-1] + ta
            p = p-1
            m = m-1
            
            
        if p > 0 and a[p][m] == a[p-1][m]+score(s[p-1],'-'):
            sa = s[p-1] + sa 
            ta = '-' + ta
            p = p-1
           
            
        if m > 0 and a[p][m] == a[p][m-1]+score('-',t[m-1]):
            sa = '-' + sa
            ta = t[m-1] + ta
            m = m-1
        return m

with open("/Users/vvpetrov/Desktop/Python_Bio/output.txt", 'w') as out:
    with open ("/Users/vvpetrov/Desktop/Python_Bio/small.fastq") as f1:
        l = ''
        l_name = ''
        l_qual = ''
        count = 0
        for line in f1:
            if count == 0: 
                l_name = line.strip()
                count += 1
                continue
            if count == 1:
                l = line.strip()
                count +=1
                continue
            if count == 2:
                count += 1
                continue
            if count == 3:
                l_qual = line.strip()
                count = 0
                coord = align(refl, l)
                print(l_name, ref_name, coord, l, l_qual, sep='\t', file=out)
            
            
    
#     print (l_name, file = bioinf.txt \t)    
#     print (ref_name,  file = bioinf.txt \t)
#     print (x, file = bioinf.txt \t)
#     print (ta, file = bioinf.txt \t)
#     print (l_qual, file = bioinf.txt \t )         
         
        
#         
```
